```python
import os
import pandas as pd
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS, Pinecone, Weaviate
from langchain.document_loaders import CSVLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.chat_models import ChatOpenAI
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain_pinecone import PineconeVectorStore
import weaviate
from langchain.retrievers import MultiQueryRetriever, EnsembleRetriever
from langchain_community.vectorstores import Weaviate
from dotenv import load_dotenv
```

```python
load_dotenv()

openai_api_key = os.getenv("OPENAI_API_KEY")
pinecone_api_key = os.getenv("PINECONE_API_KEY")
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
list_data = chosen_data
```

```python
faiss_store = None
pinecone_store = None
```

```python
file_paths = ["df_skin.csv", "df_hair.csv", "df_vits_supp.csv"]
```

```python
namespaces = {"df_skin.csv" : "df_skin", "df_hair.csv" : "df_hair", "df_vits_supp.csv" : "df_vits_supp"}
```

```python
faiss_docs = []
pinecone_docs = {"df_skin" : [], "df_hair" : [], "df_vits_supp" : []}
    
```

```python
for file in file_paths:
    loader = CSVLoader(file_path = file)
    documents = loader.load()

    text_splitter =  RecursiveCharacterTextSplitter(chunk_size = 500, chunk_overlap = 100)
    split_docs = text_splitter.split_documents(documents)

    namespace = namespaces[file]

    for doc in split_docs:
        doc.metadata["Category"] = namespace
        faiss_docs.append(doc)
        pinecone_docs[namespace].append(doc)

    print(f"Number of documents in {namespace} : {len(split_docs)}")

    print(f"Number of documents in Faiss : {len(faiss_docs)}")
```

```python
faiss_store = FAISS.from_documents(faiss_docs, embeddings)
```

```python
pinecone_store = {}
for namespace, docs in pinecone_docs.items():
    if docs:
        vector_store = PineconeVectorStore(index_name = "clinikally", embedding = embeddings, text_key = "text", namespace = namespace)
        vector_store.add_documents(docs)
        pinecone_store[namespace] = vector_store
    
```

```python
def select_retrievers(query):
    selected_retrievers = []
    for category, store in pinecone_store.items():
        if category in query.lower():
            selected_retrievers.append(store.as_retriever())
    
    if not selected_retrievers:
        selected_retrievers.append(faiss_store.as_retriever())  # Default to Pinecone if no match
    
    return selected_retrievers

# Initialize OpenAI LLM
llm = ChatOpenAI(model = "gpt-4o", openai_api_key = openai_api_key)

# Initialize memory
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# Query the system
query = "haircare products  under 1000"
retrievers = select_retrievers(query)

# Use MultiQueryRetriever to aggregate results
multi_retriever = EnsembleRetriever(retrievers = retrievers)

retrieved_docs = multi_retriever.get_relevant_documents(query)
print(retrieved_docs)

# Create QA chain with combined retriever
qa_chain = ConversationalRetrievalChain.from_llm(llm=llm, retriever=multi_retriever, memory = memory)
response = qa_chain.run(query)
print(response)
```

```python
import os
import pandas as pd
from typing import List, Dict, Any, Optional, Union

# LangChain imports - updated for latest version
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain_community.embeddings import HuggingFaceEmbeddings, OllamaEmbeddings, CohereEmbeddings, SentenceTransformerEmbeddings
from langchain_community.vectorstores import FAISS, Pinecone
from langchain_community.llms import Ollama
from langchain_community.tools import BraveSearch
from langchain_community.utilities.google_search import GoogleSearchAPIWrapper
from langchain.chains import RetrievalQA
from langchain.schema import Document
from langchain.prompts import PromptTemplate
from langchain.output_parsers import PydanticOutputParser
from pydantic import BaseModel, Field
from langchain.embeddings.base import Embeddings
from dotenv import load_dotenv
```

```python
class QueryClassifier(BaseModel):
    """Classify the query into a category."""   
    category: str = Field(description="The category of the query: 'skin care', 'hair care', 'vitamins & supplements', 'general', 'multi'")
    domains: List[str] = Field(description="The domains of the query: 'skincare', 'haircare', 'vitamins & supplements'")
    explanation: str = Field(description = "Explanation of why this classification was chosen")
```

```python
class VectorDBFactory:
    """Factory for creating vector databases."""
    
    @staticmethod
    def create_faiss(embeddings, texts: List[Document], **kwargs):
        """Create a FAISS vector database."""
        return FAISS.from_documents(texts, embeddings, **kwargs)
    
    @staticmethod
    def create_pinecone(embeddings, texts: List[Document] = None, namespace: str = None, **kwargs): #type: ignore
        """Create a Pinecone vector database."""
        from pinecone import Pinecone, ServerlessSpec

        pc = Pinecone(
            api_key = pinecone_api_key,
            environment = kwargs.get("environment", "us-east1-gcp")
        )

        index_name = kwargs.get("index_name", "clinikally-2")
        dimension = kwargs.get("dimension", 1536)

        if index_name not in pc.list_indexes().names():
            pc.create_index(
                name = index_name, 
                dimension = dimension,
                metric = "cosine",
                spec = ServerlessSpec(
                    cloud = "aws",
                    region = "us-east-1"
                )
            )
        
        index = pc.Index(index_name)

        from langchain_pinecone import PineconeVectorStore


        pinecone_store = PineconeVectorStore(
            index_name = index_name,
            embedding = embeddings, 
            namespace = namespace
        )

        if texts and namespace:
            pinecone_store.add_documents(texts, namespace = namespace)
        
        return pinecone_store
```

```python
class EmbeddingFactory:
    """Factory for creating embedding models."""

    @staticmethod
    def create_embeddings(embedding_type:str, **kwargs):
        """Create embeddings based on type"""
        if embedding_type.lower() == "openai":
            return OpenAIEmbeddings(**kwargs)
        elif embedding_type.lower() == "huggingface":
            return HuggingFaceEmbeddings(**kwargs)
        elif embedding_type.lower() == "sentence-transformer":
            return SentenceTransformerEmbeddings(**kwargs)
        else:
            raise ValueError(f"Unsupported embedding type: {embedding_type}")
```

```python
class LLMFactory:
    """Factory for creating LLM models."""

    @staticmethod
    def create_llm(llm_type:str, **kwargs):
        """Create an LLM based on type"""
        if llm_type.lower() == "openai":
            return ChatOpenAI(**kwargs)
        elif llm_type.lower() == "ollama":
            return Ollama(**kwargs)
        else:
            raise ValueError(f"Unsupported LLM type: {llm_type}")
```

```python
class SearchAPIFactory:
    """Factory for creating search APIs."""

    @staticmethod
    def create_search_api(search_type:str, **kwargs):
        """Create a search API based on type."""
        if search_type.lower() == "google":
            return GoogleSearchAPIWrapper(google_api_key = google_api_key, google_cse_id = google_cse_id, **kwargs)
        elif search_type.lower() == "brave":
            return BraveSearch.from_api_key(api_key = brave_api_key, **kwargs)
        else:
            raise ValueError(f"Unsupported search type: {search_type}")
```

```python
class DataFrameToDocumentsConverter:
    """Utility to convert a pandas DataFrame to a list of documents objects"""

    @staticmethod
    def convert(df: pd.DataFrame, text_column: str, metadata_columns: List[str]):
        documents = []
        for _, row in df.iterrows():
            metadata = {}
            if metadata_columns:
                for col in metadata_columns:
                    if col in row.index:
                        metadata[col] = row[col]
            documents.append(Document(
                page_content = str(row[text_column]),
                metadata = metadata
            ))
        return documents
```

```python
class RAGPipeline:
    """Main class for running the RAG pipeline"""

    def __init__(
            self,
            vector_db_type: str = "faiss",
            embedding_type: str = "sentence-transformer",
            llm_type: str = "openai",
            search_api_type:str = "brave",
            vector_db_kwargs: Dict[str,Any] = {}, #type: ignore
            embedding_kwargs: Dict[str, Any] = {}, #type: ignore
            llm_kwargs: Dict[str, Any] = {}, #type: ignore
            search_api_kwargs: Dict[str, Any] = {}, #type: ignore
            use_namespaces: bool = True
    ):
        self.vector_db_type = vector_db_type
        self.embedding_type = embedding_type
        self.llm_type = llm_type
        self.search_api_type = search_api_type
        self.use_namespaces = use_namespaces

        self.vector_db_kwargs = vector_db_kwargs
        self.embedding_kwargs = embedding_kwargs or {"model_name": "all-MiniLM-L6-v2"}
        self.llm_kwargs = llm_kwargs
        self.search_api_kwargs = search_api_kwargs

        self.embeddings = EmbeddingFactory.create_embeddings(embedding_type = embedding_type, **self.embedding_kwargs)

        self.llm = LLMFactory.create_llm(llm_type = llm_type, **self.llm_kwargs)

        self.domain_stores = {}
        self.domain_retrievers = {}

        try:
            self.search_api = SearchAPIFactory.create_search_api(search_type = search_api_type, **self.search_api_kwargs)
        except Exception as e:
            print(f"Warning: Search API initialization failed: {e}")
            self.search_api = None

        if vector_db_type.lower() == 'pinecone' and use_namespaces:
            if "dimension" not in self.vector_db_kwargs:
                if hasattr(self.embeddings, "embed_documents"):
                    sample_embedding = self.embeddings.embed_documents(["This is a test"])
                    if sample_embedding and len(sample_embedding) > 0:
                        self.vector_db_kwargs["dimension"] = len(sample_embedding)
            
            self.pinecone_store = VectorDBFactory.create_pinecone(
                embeddings = self.embeddings,
                **self.vector_db_kwargs
            )
    
    def add_domain_knowledge(self, domain: str, df: pd.DataFrame, text_column: str, metadata_columns: List[str] = None):
        """Add domain knowledge to the pipeline"""
        documents = DataFrameToDocumentsConverter.convert(df = df, text_column = text_column, metadata_columns = metadata_columns)

        for doc in documents:
            doc.metadata["domain"] = domain
        
        if self.vector_db_type.lower() == "pinecone" and self.use_namespaces:
            self.pinecone_store.add_documents(documents, namespace = domain)

            retriever = self.pinecone_store.as_retriever(
                search_kwargs = {
                    "k" : 4,
                    "namespace" : domain
                }
            )
            self.domain_retrievers[domain] = retriever
            self.domain_stores[domain]  = self.pinecone_store

        else:
            vector_store = None
            if self.vector_db_type.lower() == "faiss":
                vector_store = VectorDBFactory.create_faiss(
                    embeddings = self.embeddings,
                    texts = documents,
                    **self.vector_db_kwargs
                )
            elif self.vector_db_type.lower() == 'pinecone':
                vector_store = VectorDBFactory.create_pinecone(
                    embeddings = self.embeddings,
                    texts = documents,
                    **self.vector_db_kwargs
                )
            self.domain_stores[domain] = vector_store
            self.domain_retrievers[domain] = vector_store.as_retriever(search_kwargs = {"k" : 4})

    def build_classifier_chain(self):
        """Build a chain to classify user queries"""
        classifier_prompt = PromptTemplate(
            template="""You are a query classifier that categorizes user questions into domains.
            Based on the following query, determine which category it belongs to:
            - 'skin': If the query relates to skincare, skin problems, or skin health
            - 'hair': If the query relates to hair care, hair problems, or hair health
            - 'vitamins': If the query relates to vitamins, supplements, or nutrition
            - 'general': If the query is general knowledge not related to the specific domains
            - 'multi': If the query spans multiple domains

            Also identify all domains the query relates to as a list.

            Query: {query}

            {format_instructions}
            """,
            input_variables=["query"],
            partial_variables={"format_instructions": PydanticOutputParser(pydantic_object=QueryClassifier).get_format_instructions()}
        )

        classifier_chain = classifier_prompt | self.llm | PydanticOutputParser(pydantic_object = QueryClassifier)
        return classifier_chain  
    
    def process_query(self, query:str):
        """Process a user query through the pipeline"""
        print("[Line 1] Processing query:", query) # Debug line 1
        
        classifier_chain = self.build_classifier_chain()
        print("[Line 4] Built classifier chain") # Debug line 4
        
        classification = classifier_chain.invoke({"query": query})
        print("[Line 7] Classification result:", classification) # Debug line 7

        results = {
            "classification": classification,
            "sources": [],
            "answer": ""
        }
        print("[Line 14] Initialized results dict") # Debug line 14

        if classification.category == "multi":
            print("[Line 16] Handling multi-domain query") # Debug line 16
            combined_documents = []

            for domain in classification.domains:
                print(f"[Line 19] Processing domain: {domain}") # Debug line 19
                if domain in self.domain_retrievers:
                    domain_docs = self.domain_retrievers[domain].invoke(query)
                    print(f"[Line 22] Retrieved {len(domain_docs)} docs for domain {domain}") # Debug line 22

                    for doc in domain_docs:
                        if "domain" not in doc.metadata:
                            doc.metadata["domain"] = domain
                    
                    combined_documents.extend(domain_docs)
                
            if combined_documents:
                print(f"[Line 31] Total combined documents: {len(combined_documents)}") # Debug line 31
                qa_prompt = PromptTemplate(
                    template="""You are a helpful assistant answering questions about skin care, hair care, and vitamins.
                    Use the following retrieved documents to answer the user's question.
                    If you don't know the answer, just say that you don't know.
                    If information comes from multiple domains, make it clear in your answer.
                    
                    Context: {context}
                    
                    Question: {question}
                    
                    Answer:""",
                    input_variables=["context", "question"]
                )

                context = "\n\n".join([f"[{doc.metadata.get('domain', 'unknown')}] {doc.page_content}" for doc in combined_documents])
                print("[Line 49] Built context from combined documents") # Debug line 49
                
                # Generate answer
                results["answer"] = self.llm.invoke(qa_prompt.format(context=context, question=query))
                print("[Line 52] Generated answer for multi-domain query") # Debug line 52
                print("we are here and the answer is ",results["answer"])
                results["sources"] = combined_documents
            
            else:
                results["answer"] = "I couldn't find relevant information in my knowledge base for your multi-domain query."
                print("[Line 57] No documents found for multi-domain query") # Debug line 57
    
        elif classification.category in ["skin", "hair", "vitamins"]:
            print(f"[Line 60] Handling single domain query for: {classification.category}") # Debug line 60
            domain = classification.category
            if domain in self.domain_retrievers:
                retriever = self.domain_retrievers[domain]
                print("th")
                qa_chain = RetrievalQA.from_chain_type(
                    llm = self.llm,
                    retriever = retriever,
                    chain_type = "stuff",
                    return_source_documents = True
                )
                print("[Line 69] Created QA chain for single domain") # Debug line 69

                qa_result = qa_chain.invoke({"query": query})
                print("[Line 72] Got QA result for single domain")
                print("this is the qa result", qa_result) # Debug line 72
                results["answer"] = qa_result["result"]
                results["sources"] = qa_result["source_documents"]
            
            else:
                results["answer"] = f"I couldn't find relevant information specific to {domain} in my knowledge base for your query."
                print(f"[Line 78] No retriever found for domain: {domain}") # Debug line 78
        
        elif classification.category == "general":
            print("[Line 81] Handling general query") # Debug line 81
            if self.search_api:
                try: 
                    search_results = self.search_api.invoke(query)
                    print("[Line 85] Got search results") # Debug line 85
                    results['sources'] = search_results
                
                    if isinstance(search_results, list):
                        context = "\n\n".join([str(result) for result in search_results])
                    else:
                        context = str(search_results)
                    print("[Line 92] Built context from search results") # Debug line 92
                    
                    qa_prompt = PromptTemplate(
                        template="""You are a helpful assistant.
                        Use the following search results to answer the user's question.
                        
                        Search Results: {context}
                        
                        Question: {question}
                        
                        Answer:""",
                        input_variables=["context", "question"]
                    )
                    results["answer"] = self.llm.invoke(qa_prompt.format(context=context, question=query))
                    print("[Line 106] Generated answer for general query") # Debug line 106
                except Exception as e:
                    print(f"[Line 108] Error in search API: {str(e)}") # Debug line 108
                    results["answer"] = f"I tried to search for information, but encountered an error: {str(e)}"
            else:
                results["answer"] = "I don't have access to a search API to look up general information."
                print("[Line 112] No search API available") # Debug line 112
```
